```python
def quick_sort_func(Array,Start,End):
    if(Start < End):
        i = Start
        x = Start
        j = End
        while(i < j):
            while(Array[i] <= Array[x] and i < End):
                i = i+1
            while(Array[j] > Array[x]):
                j = j-1
            if(i < j):
                temp = Array[i]
                Array[i] = Array[j]
                Array[j] = temp
        temp = Array[x]
        Array[x] = Array[j]
        Array[j] = temp
        quick_sort_func(Array,Start,j-1)
        quick_sort_func(Array,j+1,End)
```

```python
def Floyd_Warshall_Algorithm_with_deadline(V,graph,time,time_required):
    min_int = -np.inf
    cost = np.array(graph)
    time_taken = 0
    for i in range(0,V):
        for j in range(0,V):
            cost[i][j] = graph[i][j]
    for k in range(0,V):
        for u in range(0,V):
            for v in range(0,V):
                if(cost[u][k] != min_int and cost[k][v] != min_int and cost[u][v] < cost[u][k] + cost[k][v] and time_required[u][k] != -1 and time_required[k][v] != -1 and time_taken + time_required[u][k] + time_required[k][v] <= time):
                    cost[u][v] = cost[u][k] + cost[k][v]
                    time_taken = time_taken + time_required[u][k] + time_required[k][v]
                    
    return cost
```
